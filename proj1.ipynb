```python
import pandas as pd

df = pd.read_csv('survey_results_public.csv')
print(df.head())
```

```python
df.set_index('ResponseId', inplace=True)

```

```python
spec_data = df[['RemoteWork', 'MainBranch', 'Age']]


print(spec_data.head())

```

```python
filter = (df['Q120'] == 'I agree') & (df['Age'] == '25-34 years old')
filtered_data = df[filter]

print(filtered_data)
```

```python
sorted_data = df.sort_values(by=['Q120', 'MainBranch'], ascending=[True, True])
print(sorted_data.head())
```

```python
print(df['MainBranch'].value_counts())
```

```python
import matplotlib.pyplot as plt
        
```

```python
age_distribution = df['Age'].value_counts().sort_index()
age_distribution.plot(kind='bar')
plt.title('Age Distribution of Respondents')
plt.xlabel('Age')
plt.ylabel('Number of Respondents')
plt.show()
```

```python
median_compensation_by_age = df.groupby('Age')['ConvertedCompYearly'].median().dropna()

# Plot the line chart
median_compensation_by_age.plot(kind='line', marker='o', linestyle='-')
plt.title('Median Yearly Compensation by Age')
plt.xlabel('Age')
plt.ylabel('Median Yearly Compensation')
plt.xticks(rotation=45)  
plt.grid(True)  
plt.tight_layout() 
plt.show()
```
